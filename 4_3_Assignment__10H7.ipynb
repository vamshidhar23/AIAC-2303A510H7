```python
#Task 1: Zero-Shot Prompting – Leap Year Check
def is_leap_year(year):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        return "Leap Year"
    else:
        return "Not a Leap Year"


year = int(input("Enter a year: "))
print(is_leap_year(year))
```

```python
#Task 2: One-Shot Prompting – Centimeters to Inches
def cm_to_inches(cm):
    return cm / 2.54


cm = float(input("Enter length in centimeters: "))
print("Length in inches:", round(cm_to_inches(cm), 2))
```

```python
#Task 3: Few-Shot Prompting – Name Formatting
def format_name(full_name):
    parts = full_name.split()
    first_name = parts[0]
    last_name = parts[-1]
    return last_name + ", " + first_name


name = input("Enter full name: ")
print(format_name(name))
```

```python
#Task 4: Comparative Analysis – Zero-Shot vs Few-Shot (Vowel Count)
def count_vowels_zero_shot(text):
    count = 0
    for ch in text.lower():
        if ch in "aeiou":
            count += 1
    return count


text = input("Enter a string: ")
print("Vowel count:", count_vowels_zero_shot(text))
def count_vowels_few_shot(text):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in text if ch in vowels)


text = input("Enter a string: ")
print("Vowel count:", count_vowels_few_shot(text))
```
